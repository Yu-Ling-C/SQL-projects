6. Display the top 10 drivers ranked by their number of pole positions (`driverStandings.position` equals 1)

```python
select 
	d.forename,
	d.surname,
	count(ds.position) as num_of_1   # count how many times each driver is ranked number 1 in 'driverStandings'
from drivers d 
LEFT JOIN driverStandings ds
USING(driverId)
WHERE ds.position = 1
group by 1,2 
order by 3 DESC
limit 10  
```

7. How many times did Ayrton Senna won a race? How many times did he finish second?

```python
SELECT d.surname,
	(select count(r.position)   
	from results r  
	left join drivers d 
	USING(driverId)
	WHERE r.position = 1 and d.surname = 'Senna' and d.forename = 'Ayrton'
	group by d.surname) as num_win,     # subquery: count how many times Senna won (when r.position = 1)
	(select count(r.position) 
	from results r
	left join drivers d 
	USING(driverId)
	WHERE r.position = 2 and d.surname = 'Senna' and d.forename = 'Ayrton'
	group by d.surname) as num_of_2     # subquery: count how many times Senna finished second (when ds.position = 2)
from drivers d
WHERE d.surname = 'Senna' and d.forename = 'Ayrton'   

# Ayrton Senna won first place in 41 races and finished second for 23 times.
```

|surname|num_win|num_of_2|
|-------|-------|--------|
|Senna  |41     |23      |

8. Display the top 10 constructors ranked by the number of victories of their pilots

```python
# First define the number of victories of their pilots (where r.position = 1)
# find connection between two tables (using 'results' to connect 'drivers' and 'constructors' tables)

SELECT 
	c.name as constructor_name, 
	COUNT(r.position) as pilot_num_vic   # count the total number of victory based on constructors when their drivers are ranked number 1
FROM constructors c 
left join results r 
USING(constructorId)
left join drivers d
USING(driverId)
where r.position = 1
group by constructor_name 
order BY pilot_num_vic DESC 
limit 10
```

10. With which constructor did Juan Fangio won the highest number of Grand Prix?

```python
SELECT 
	d.forename, 
	d.surname,
	COUNT(r.position) as pilot_num_vic,   # count the number of victory when drivers are ranked number 1 in 'results'
	c.name as constructor_name
FROM drivers d 
left join results r 
USING(driverId)
left join constructors c 
USING(constructorId)
WHERE r.position = 1 AND d.forename = 'Juan' and d.surname = 'Fangio'  # filter the driver 'Juan Fangio'
group BY 1,2,4
order BY 3 DESC 

# By looking at the result table, we know that Juan Fangio won the highest number of Grand Prix with Mercedes
```
